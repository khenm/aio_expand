# Broadcasting: The Unsung Hero
Making Arrays with mismatched shapes compatible for arithmetic operations. 

It "extends" smaller arrays to the shape of the larger array

Element-wise operation on arrays of different shapes

```python
# broadcasting and vectorization
a = np.array([[1, 2], [3, 4], [5, 6]])
b = np.array([1, 2])

# broadcasting in action: adding the 1D array b to each row of array a
print(a + b)
```

# Vectorization: Powering Efficient Computations
The execution of operations on entire arrays, eliminating the need for explicit loops.

Making calculations significantly faster than traditional Python loops

Almost all NumPy operations are inherently vectorized


```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

print(a * b) # element wise multiplication
```

```python
import time
def mutiply_list(size=1000):
    list1 = list(range(size))
    list2 = list1.copy()
    
    initial_time = time.time()
    
    result = [(a * b) for a, b in zip(list1, list2)]
    
    print("Time taken:", time.time() - initial_time)
    
def multiply_array(size=1000):
    arr1 = np.array(list(range(size)))
    arr2 = arr1.copy()
    
    initial_time = time.time()

    result = arr1 * arr2

    print("Time taken:", time.time() - initial_time)
    
mutiply_list()
multiply_array()
```

### Resolving brightness overflow
We know that if we plus over the limit of the uint8, we will get the value from start, so we have to use clip(). \
Clip allows that number to be shown. 

```python
data = np.array([2]).astype(float)
data = data - 100
data = np.clip(data, a_min = 0, a_max = 255)
data = data.astype(np.uint8)
print(data)
```

Another solution is to use where() function

```python
# create array
arr = np.arange(5)
print(arr)

condition = arr < 3
out = np.where(condition, arr, arr * 2)

print(condition)
print(out)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read data
data = pd.read_csv('advertising.csv')
sales = np.array(data['Sales'])
radio = np.array(data['Radio'])

ones = np.repeat(1, radio.size).reshape(radio.size, 1)
radio_stack = np.hstack((radio.reshape(radio.size, 1), ones))
sales = sales.reshape(sales.size, 1)
```

```python
# least square estimations
import numpy as np
ata = radio_stack.T @ radio_stack
coef = np.linalg.inv(ata) @ radio_stack.T @ sales

estimation_results = radio_stack @ coef

plt.figure(figsize=(10, 6))  # Set the figure size

# Plot the data points
plt.scatter(radio, sales, color='blue', label='Data Points')

plt.plot(radio, estimation_results, color='red', label='Fit Line')

data = pd.read_csv('advertising.csv')
sales = data['Sales']
radio = data['Radio']

coefficients = np.polyfit(radio, sales, 1)
polynomial = np.poly1d(coefficients)
line = polynomial(radio)

plt.plot(radio, line, color='green', label='Built-in Fit Line')

# Add title and labels
plt.title('Sales vs. Radio Advertising')
plt.xlabel('Radio Advertising Spend')
plt.ylabel('Sales')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# find variance
difference = np.abs(estimation_results - line)
avg = np.average(difference)
print(f'Variance average score: {round(avg, 3)}')
```
